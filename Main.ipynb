## Общий вывод

Задача состояла в разработке решения для интернет-магазина "В один клик", направленного на увеличение покупательской активности постоянных клиентов путем персонализации предложений. Основные цели включали:

1) Построение модели для прогнозирования вероятности снижения покупательской активности клиента в следующие три месяца.

2) Выделение сегментов покупателей и разработку персонализированных предложений на основе данных модели и информации о прибыльности клиентов.

Исходные данные были представлены несколькими файлами:

market_file.csv: данные о поведении покупателей на сайте, коммуникациях и продуктовом поведении.
market_money.csv: данные о выручке с покупателей.
market_time.csv: данные о времени, проведенном на сайте.
money.csv: данные о среднемесячной прибыли покупателей.

Предобработка данных включала:

Проверка данных на наличие пропусков и дубликатов
Исправление опечатки в переменной "Тип сервиса".
Агрегирование данных о выручке и времени по периодам.
Объединение всех таблиц в единый датасет.
Удаление выброса (клиент с аномально высокой выручкой).
Создание дополнительных признаков, таких как разница в выручке между периодами и ошибки сайта на минуту пользования.


Для поиска лучшей модели были выполнены следующие шаги:

Создан пайплайн для предобработки данных, включающий обработку числовых и категориальных признаков.
Использован GridSearchCV для перебора гиперпараметров следующих моделей:

DecisionTreeClassifier
KNeighborsClassifier
LogisticRegression
SVC


Для каждой модели были подобраны различные гиперпараметры и методы масштабирования данных.
Оценка моделей производилась с использованием метрики ROC AUC.


Выводы и предложения для выбранного сегмента:
Был выделен сегмент покупателей с высокой вероятностью снижения активности (выше 0.8) и высокой прибыльностью (больше 4). Для этого сегмента были сделаны следующие выводы и предложения:

1) Увеличить количество маркетинговых коммуникаций, так как эти пользователи реже получают рекламные сообщения, но не против их получать.

2) Сосредоточиться на устранении ошибок сервиса, характерных для этого сегмента. После исправления ошибок предложить пользователям скидки в качестве компенсации.

3) Разработать стратегию для товаров в корзине, которые не были оплачены. Предложить скидки на отобранные товары или улучшить инструмент сравнения товаров.

4) Увеличить количество акционных предложений для этого сегмента, так как они склонны к покупкам по акциям и остаются прибыльными.

5) Сфокусироваться на улучшении предложений в категориях косметики, кухонной утвари и товаров для детей, которые наиболее популярны в этом сегменте.

### Вывод информации о датафрейме

```python
def df_look(df):
    """
    Функция для базового анализа Датасета.
    Выводит общую информацию о датасете, включая первые строки,
    типы данных, пропущенные значения и дубликаты.
    """
    
    # Вывод первых 5 строк датафрейма
    print("----- Первые 5 наблюдений-----")
    display(df.head(5))
    
    # Вывод основной информации о датафрейме
    print('\n----- Основная информация о датасете-----')
    df.info(verbose=True, show_counts=True)
    
    # Подсчет и вывод количества пропущенных значений
    print("\n----- Количество пропущенных значений -----")
    display(df.isnull().sum())
    
    # Проверка на наличие полных дубликатов
    print("\n ----- Наличие дубликатов -----")
    display(df.duplicated().sum())
    
    # Проверка на наличие повторяющихся id, если такой столбец есть
    if 'id' in df.columns:
        duplicates = df.duplicated(subset=['id'], keep=False)
        print(f"Количество повторяющихся id: {duplicates.sum()}")
    else:
        print("\nПредупреждение: столбец 'id' отсутствует в датафрейме.")
```

### Функция для визуализации данных

```python
def vis_inspect(data, var, bins='auto'):
    """
    Функция для визуализации основных описательных статистик.
    Для количественной переменной выводит гистограмму и боксплот. 
    Для категориальной столбчатую диаграмму
    """
    if data[var].dtype == 'object' or data[var].dtype.name == 'category':
        plt.figure(figsize=(15, 10))
        
        # Подсчет значений и процентов
        value_counts = data[var].value_counts()
        total = len(data[var])
        percentages = value_counts / total * 100
        
        # Создание барплота
        ax = sns.barplot(x=value_counts.index, y=value_counts.values)
        
        # Добавление подписей с процентами
        for i, p in enumerate(ax.patches):
            height = p.get_height()
            ax.text(p.get_x() + p.get_width()/2., height,
                    f'{height}\n({percentages[i]:.1f}%)',
                    ha='center', va='bottom', fontsize=12)
        
        plt.title(f'Барплот переменной {data[var].name}', fontsize=18)
        plt.xlabel('Категории', fontsize=16)
        plt.ylabel('Количество', fontsize=16)
        plt.xticks(rotation=45, ha='right', fontsize=12)
        plt.yticks(fontsize=12)
        plt.tight_layout()
        plt.show()
    else:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x=var, kde=True, bins=bins)
        plt.title(f'Гистограмма переменной {data[var].name}', fontsize=18)
        plt.xlabel('Значение', fontsize=16)
        plt.ylabel('Частота', fontsize=16)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.show()
        
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=data, x=var)
        plt.title(f'Боксплот переменной {data[var].name}', fontsize=18)
        plt.xlabel('Значение', fontsize=16)
        
        # Добавление подписей к боксплоту
        q1, median, q3 = data[var].quantile([0.25, 0.5, 0.75])
        min_val = data[var].min()
        max_val = data[var].max()
        mean_val = data[var].mean()
        
        plt.text(min_val, -0.15, f'Min\n{min_val:.2f}', ha='center', va='center', fontsize=12)
        plt.text(q1, -0.15, f'Q1\n{q1:.2f}', ha='center', va='center', fontsize=12)
        plt.text(median, -0.15, f'Median\n{median:.2f}', ha='center', va='center', fontsize=12)
        plt.text(q3, -0.15, f'Q3\n{q3:.2f}', ha='center', va='center', fontsize=12)
        plt.text(max_val, -0.15, f'Max\n{max_val:.2f}', ha='center', va='center', fontsize=12)
        
        # Добавление точки со средним значением
        plt.plot(mean_val, 0, 'ro', markersize=10)
        plt.text(mean_val, 0.15, f'Mean\n{mean_val:.2f}', ha='center', va='center', color='red', fontsize=12)
        
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.tight_layout()
        plt.show()
```

### Функция для построения stacked bat plot и изучения влияния переменной на целевой признак

```python
def chi_stacked(data, x, y, title="Связь между независимой и целевой переменной"):
    """
    Функция для анализа связи между независимыми и целевой переменной с помощью столбчатой диаграммы и стастистики хи-квадат.
    Количественную переменную функция разбивает на 10 равнонаполненных корзин.
    """
    
    
    
    # Получаем независимую и зависимую переменные из датафрейма
    independent_var = data[x]
    dependent_var = data[y]
    
    # Проверяем, является ли независимая переменная количественной
    if np.issubdtype(independent_var.dtype, np.number):
        # Разделяем количественную переменную на 10 равных частей
        independent_var = pd.qcut(independent_var, q=10, duplicates='drop')
    
    # Создаем кросс-таблицу
    cont_table = pd.crosstab(independent_var, dependent_var)
    
    # Вычисляем пропорции
    props = cont_table.div(cont_table.sum(axis=1), axis=0)
    
    # Создаем мозаичный график
    fig, ax = plt.subplots(figsize=(12, 6))
    props.plot(kind='bar', stacked=True, ax=ax)
    
    ax.set_title(title)
    ax.set_xlabel(x)
    ax.set_ylabel("Пропорция")
    ax.legend(title=y)
    
    # Добавляем подписи процентов вручную
    for i, c in enumerate(ax.containers):
        heights = [p.get_height() for p in c]
        labels = [f'{h:.2f}%' for h in heights]
        
        for j, (rect, label) in enumerate(zip(c, labels)):
            height = rect.get_height()
            if i == 0:  # bottom bar
                y_pos = height / 2
            else:
                y_pos = rect.get_y() + height / 2
            x_pos = rect.get_x() + rect.get_width() / 2
            ax.text(x_pos, y_pos, label, ha='center', va='center')
    
    # Выполняем хи-квадрат тест
    chi2, p_value, dof, expected = chi2_contingency(cont_table)
    
    plt.tight_layout()
    plt.show()
    
    print(f"Chi-squared statistic: {chi2:.3f}")
    print(f"p-value: {p_value:.3f}")
```

## Разработка пайплана и базовая модель
Используем модель логистической регрессиии со всеми доступными предикторами

Базовая модель показывает неплохой результат и позволяет предсказать класс клиента с точностью 87%. 
Базовая модель лучше работает с отрицательным классом (сохранение покупательной активности) нежели с предсказанием снижения.
Текущая модель пропускает 23% интересующих нас наблюдений

Метрика ROC AUC показывает точность модели 0.89.

Мы будем пользоваться именно этой метрикой по двум причинам: 
1) у нас несбалансированная целевая переменная 

2) нам нужен баланс между точностью и полнотой идентификации целевой переменной. Ложно отрицательный результат привет к тому, что мы окончательно упустим клиентов со снижающейся активностью. Ложно положительный результат приведет к лишнему расходованию ресурсов на клиентов, которые сохраняют свою активность


```python
# Задаем константы
RANDOM_STATE = 181
TEST_SIZE=0.2
```

```python
# Инициализация LabelEncoder для кодирования целевой переменной
le = LabelEncoder()

# Подготовка признаков (X) и целевой переменной (y)
X = market_data.drop(['Покупательская активность', "id"], axis=1)  # Удаляем целевую переменную и id из признаков
y = le.fit_transform(market_data['Покупательская активность'])  # Кодируем целевую переменную

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)

# Определение числовых и категориальных признаков
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Создание пайплайна для обработки числовых признаков
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Заполнение пропусков медианой
    ('scaler', StandardScaler())  # Стандартизация числовых признаков
])

# Создание пайплайна для обработки категориальных признаков
categorical_transformer = Pipeline([
    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),  # Заполнение пропусков наиболее частым значением
    ('ohe', OneHotEncoder(drop='first', handle_unknown='error', sparse_output=False))  # One-hot кодирование
])

# Объединение трансформеров для числовых и категориальных признаков
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'  # Оставляем неизменными признаки, не вошедшие в num и cat
)

# Создание полного пайплайна, включающего предобработку и модель
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('models', LogisticRegression())  # Используем логистическую регрессию как модель
])
```

```python
# Обучение модели на тренировочных данных
model.fit(X_train, y_train)

# Предсказание классов для тестовых данных
y_pred = model.predict(X_test)

# Получение вероятностей для положительного класса
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Вывод отчета о качестве классификации
print(classification_report(y_test, y_pred))
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC AUC score: {roc_auc}")
```

## Оптимизация модели и подбор гиперпараметров

Оптимизация модели с помощью перебора техник классификации и гиперпараметров позволили улучшить точность модели до 0.91

По итогам оптимизации наилучшие результаты показала Логистическая регрессия с L1-регуляризацией (Лассо), параметром регуляризации C=10, методом решения 'liblinear', масштабированием числовых признаков с помощью MinMaxScaler

```python
param_grid = [
    # Словарь для DecisionTreeClassifier
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 7),  # Перебор глубины дерева от 2 до 5
        'models__max_features': range(2, 5),  # Перебор максимального числа признаков от 2 до 4
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), None]  # Варианты масштабирования числовых признаков
    },
    
    # Словарь для KNeighborsClassifier
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 11),  # Перебор числа соседей от 2 до 9
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), None]  # Варианты масштабирования числовых признаков
    },
    
    # Словарь для LogisticRegression
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear', penalty='l1')],
        'models__C': [0.1, 1, 10, 100],  # Перебор параметра регуляризации
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), None]  # Варианты масштабирования числовых признаков
    },
    
    # Словарь для SVC 
    {
        'models': [SVC(random_state=RANDOM_STATE)],
        'models__C': [0.1, 1, 10],  # Перебор параметра регуляризации
        'models__kernel': ['linear'],  # Использование только линейного ядра
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler()]  # Варианты масштабирования числовых признаков
    }
]

# Настройка и запуск GridSearchCV
grid_search = GridSearchCV(
    model,
    param_grid,  
    cv=5,  
    scoring='roc_auc', 
    n_jobs=-1  
)

# Обучение GridSearchCV на тренировочных данных
grid_search.fit(X_train, y_train)
```
